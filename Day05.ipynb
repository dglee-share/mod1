```python
with open('전국렌터카업체정보표준데이터.csv','r') as file:
    raw = file.readlines()

#print(raw)

raw_split=[]
for line in raw:
    raw_split.append(line.split(','))
#print(raw_split)

#for index, name in enumerate(raw_split[0]):
#    print(index, name)

company_address=[]
for company in raw_split[1:]:
    company_address.append(company[2])

#print(company_address)
address_heading=[]
for address in company_address:
    try:
        address_heading.append(address.split()[0].lstrip('"'))
    except:
        pass
    
print(address_heading)
```

```python
address_set = set(address_heading)
print(address_set)
```

```python
for location in address_set:
    n = address_heading.count(location)
    print(location, n)
```
